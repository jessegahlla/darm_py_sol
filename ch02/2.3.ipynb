Demonstration of the Central Limit Theorem: let $x = x_1 + ... + x_{20}$, the sum of 20 independent Uniform(0,1) random variables. In ~~R~~ Python, create 1000 simulations of $x$ and plot their histogram. On the histogram, overlay a graph of the normal density function. Comment on any differences between the histogram and the curve.

```python
# Single uniform random variable between 0 and 1

import numpy as np
import matplotlib.pyplot as plt

n_simulations = 1000

random_list = list()

for random_variable in range(n_simulations):
    x = np.random.uniform(0,1,20).sum()
    random_list.append(x)

random_array = np.array(random_list)
```

```python
from scipy.stats import norm

x = np.linspace(norm.ppf(0.01),
               norm.ppf(0.99),
               100)
```

```python
fit = norm.pdf(random_array, np.mean(random_array), np.std(random_array))

count, bins, ignored = plt.hist(random_array, 15, normed=True, label='random sums')
plt.plot(random_array, fit, 'r.', lw=3, alpha = 0.6, label='normal pdf')
plt.xlabel('Random Variable')
plt.ylabel('Frequency')
plt.title('The history of sums of uniform random numbers are similar to a normal pdf')
plt.legend()
plt.show()
```
